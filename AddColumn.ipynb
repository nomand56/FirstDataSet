```python
import pandas as pd
```

```python
data=pd.read_csv("Pokemon.csv")
```

```python
data.head(n=10)
```

```python
datamean=data['Speed'].mean()
```

```python
datamean
```

```python
def speed_low(num):
    if num < datamean:
        return "Low Speed"
    elif num ==datamean:
        return "equal speed"
    else:
        return "low speed"
```

```python
data["High_low_Speed"]=data["Speed"].apply(speed_low)
data
```

```python
hpmean=data["HP"].mean()
def high_hp(num1):
    if num1>=hpmean:
        return "High low Speed"
    else:
        return "Low Speed"
```

```python
data["high low hp"]=data["HP"].apply(high_hp)
data
```
